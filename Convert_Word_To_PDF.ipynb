```python
import os
source_folder = "/content/project_data"  # CHANGE THIS to your actual folder
pdf_folder = os.path.join(source_folder, "pdf")  # Create a 'pdf' folder inside

# Create 'pdf' folder if it doesn't exist
if not os.path.exists(pdf_folder):
    os.makedirs(pdf_folder)

# Convert all .docx files to PDF
for file in os.listdir(source_folder):
    if file.endswith(".docx"):
        docx_path = os.path.join(source_folder, file)
        pdf_path = os.path.join(pdf_folder, file.replace(".docx", ".pdf"))

        # Convert using LibreOffice
        os.system(f"libreoffice --headless --convert-to pdf '{docx_path}' --outdir '{pdf_folder}'")
        print(f"Converted: {file} -> {pdf_path}")

print("✅ Batch conversion completed! PDFs are stored in:", pdf_folder)
```
